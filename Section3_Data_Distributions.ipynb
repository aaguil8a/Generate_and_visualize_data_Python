```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
## Normally-distributed numbers

# parameters

strech = 3 # variance (square of Standard Deviation)
shift = 10 # mean
n = 1000

# create data

data = strech* np.random.randn(n) + shift


# plot
fig = plt.figure(figsize=[6,6])

plt.subplot2grid((2,2), (0,0))
plt.plot(data)


plt.subplot2grid((2,2),(0,1))
plt.hist(data,20)

plt.subplot2grid((2,2), (1,0))
stats.probplot(data,dist='norm', plot=plt)

plt.subplot2grid((2,2), (1,1))
plt.violinplot(data,showmedians=True)
plt.show()
```

```python
## uniformly-distributed numbers

# parameters

strech = 1 # variance (square of Standard Deviation)
shift = 0 # mean
n = 1000

# create data
                     # rand gives you uniform numbers
data = np.random.rand(n) + shift-.5*strech


# plot
fig = plt.figure(figsize=[6,6])

plt.subplot2grid((2,2), (0,0))
plt.plot(data)


plt.subplot2grid((2,2),(0,1))
plt.hist(data,20)

plt.subplot2grid((2,2), (1,0))
stats.probplot(data,dist='norm', plot=plt)

plt.subplot2grid((2,2), (1,1))
plt.violinplot(data,showmedians=True)
plt.show()
```

```python
# generate data

n = 1000 

data = np.exp(np.random.randn(n) *.8) # log-norm distribution 

stats.probplot(data,dist='norm', plot=plt)
plt.show()
```

```python
# number of points 

n = 1000

# mean and variance 
# lamba parameter
l = 8

# initialize data

data = np.zeros(n)

# loop over points

for i in range(0,n):

  # reset values
  p = 1
  k = 1

  # randomly decrease p until it exceeds lamda (the threshold)

  while p > np.exp(-1):
    p = p * np.random.rand(1) #rand is [0,1] so p keeps getting smaller
    k = k + 1

    # data points for this iteration
    data[i] = k-2


fig = plt.figure(figsize=[6,6])

plt.subplot2grid((2,2), (0,0))
plt.plot(data)


plt.subplot2grid((2,2),(0,1))
plt.hist(data,50)

plt.subplot2grid((2,2), (1,0))
stats.probplot(data,dist='norm', plot=plt)

plt.subplot2grid((2,2), (1,1))
plt.violinplot(data,showmedians=True)
plt.show()



```
