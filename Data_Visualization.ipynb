```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
from datetime import datetime
import dateutil.parser as parser
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

```python
filename = '/content/Abortion Support (3).xlsx'

df = pd.read_excel(filename)

```

```python
df.rename(columns = {'Poll Date':
                     
                     'Poll_Date',

                     'Legal under any %': 'Legal_under_any_%',
                     'Legal only under certain %': 'Legal_only_under_certain_%',
                     'Illegal in all %': 'Illegal_in_all_%',
                     'No opinion %': 'No_opinion_%'}, inplace = True)
```

```python
df['Poll_Date'] = df['Poll_Date'].str.split('-', expand=True)[0]
```

```python
df.Poll_Date
```

```python
df.info()
```

```python
df['year'] = pd.DatetimeIndex(df['Poll_Date']).year
```

```python
sns.displot()
```

```python
sns.lineplot(x='year', y='Legal_only_under_certain_%', markers=True, ci=None, data=df)
plt.title('Percent of legal only under certain')
plt.ylabel('Percent:Legal Only Under Certain')
plt.xlabel('Poll Date');
```

```python
sns.lineplot(x='year', y='Legal_under_any_%',markers=True, ci=None,data=df)
plt.title('Percent that are for abortion: legal under any circumstance')
plt.ylabel('Percent: legal under any circumstance')
plt.xlabel('Poll Date')
```

```python
sns.lineplot(x='year', y='Illegal_in_all_%', markers=True, ci=None, data=df)
plt.title('Percent that are against abortion: illegal in all circumstance')
plt.ylabel('Percent: illegal in all circumstance')
plt.xlabel('Poll Date');
```

```python
sns.lineplot(x='year', y='Illegal_in_all_%', markers=True, ci=None, data=df)
plt.title('Percent that are against abortion: illegal in all circumstance')
plt.ylabel('Percent: illegal in all circumstance')
plt.xlabel('Poll Date');
```

```python
sns.lineplot(x='year', y='No_opinion_%',markers=True, ci=None, data=df)
plt.title('Percent that have no opinion')
plt.ylabel('Percent: No Opinion')
plt.xlabel('Poll Date');
```
