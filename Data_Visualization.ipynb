```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
from datetime import datetime
import dateutil.parser as parser
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

```python
filename = '/content/Abortion Support.xlsx'

df = pd.read_excel(filename, parse_dates=['Poll Date'], index_col='Poll Date')

sns.set(font_scale=1.5, style="whitegrid")
```

```python
df.head()
```

```python
df.index.values
```

```python
# Created figure and plot space
fig, ax = plt.subplots(figsize=(30, 10 ))

# Added x-axis and y-axis
ax.scatter(df.index.values,
        df['Legal under any %'],
        color='purple')

# Added title and labels for axes
ax.set(xlabel="Date",
       ylabel="Legal in any case",
       title="Americas for Abortion")

plt.setp(ax.get_xticklabels(), rotation=45)

plt.show()
```
