```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
from datetime import datetime
import dateutil.parser as parser
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

```python
filename = '/content/Abortion Support (3).xlsx'

df = pd.read_excel(filename)

```

```python
df.T
```

```python
# create the time series subplots
fig,ax =  plt.subplots( 2, 2,
                       figsize = ( 20, 8))
  
sns.lineplot( x = "Poll Date", y = "Legal under any %", 
             color = 'r', data = df, 
             ax = ax[0][0])
  
ax[0][0].tick_params(labelrotation = 30)
sns.lineplot( x = "Poll Date", y = "Legal only under certain %", 
             color = 'g', data = df,
             ax = ax[0][1])
  
ax[0][1].tick_params(labelrotation = 25)
sns.lineplot(x = "Poll Date", y = "Illegal in all %", 
             color = 'b', data = df,
             ax = ax[1][0])
  
ax[1][0].tick_params(labelrotation = 25)
  
sns.lineplot(x = "Poll Date", y = "No opinion %", 
             color = 'y', data = df, 
             ax = ax[1][1])
  
ax[1][1].tick_params(labelrotation = 25)
fig.tight_layout(pad = 1.2)

```

```python
# Created figure and plot space
fig, ax = plt.subplots(figsize=(30, 10 ))

# Added x-axis and y-axis
ax.scatter(df.index.values,
        df['Legal under any %'],
        color='purple')

# Added title and labels for axes
ax.set(xlabel="Date",
       ylabel="Legal in any case",
       title="Americas for Abortion")

plt.setp(ax.get_xticklabels(), rotation=45)

plt.show()
```
